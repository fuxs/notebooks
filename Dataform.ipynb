# Dataform

## Repository

A **repository** corresponds with a Git repository and contains:

* SQLX files
* Javascript files
* Configuraion files
* NPM packages

Each Dataform repository is connected to a service account, which requires at least the role `roles/bigquery.user`.

### Create a repository

You require the role `roles/dataform.admin` to create a new repository.

### Connect to a third-party repository

## Workspace

A workspace contains an editable copy of your team's repository.

Each table and view definition is stored in a separate file.

# Example

1. Go to **Dataform** by clicking on this [link](https://console.cloud.google.com/bigquery/dataform).

2. Create a **repository** by clicking on [+ CREATE REPOSITORY](https://console.cloud.google.com/bigquery/dataform/create).

3. Provide the repository id: ***my-repository*** and the region: ***us-central1***, keep the default service account.

4. Create a **development workspace** by clicking on [+ CREATE DEVELOPMENT WORKSPACE](https://console.cloud.google.com/bigquery/dataform/locations/us-central1/repositories/my-repository/details/workspaces?showCreateWorkspaceSubtask=true), provide the workspace id ***development***

5. Initialize the new workspace by clicking on the button [INITIALIZE WORKSPACE](https://console.cloud.google.com/bigquery/dataform/locations/us-central1/repositories/my-repository/workspaces/development/files/definitions)

6. Delete all files in the `definitions` folder.

7. Open the `workflow_settings.yaml` and check the settings. It should look like:

```yaml
defaultProject: your_projectd_id
defaultLocation: US
defaultDataset: dataform_ds
defaultAssertionDataset: dataform_assertions
dataformCoreVersion: 3.0.0
```

8. Create a new source file `definitions/patient_data.sqlx` with the following content:

```json
config {
  type:   "declaration",
  schema: "dataform_ds",  // BigQuery dataset
  name:   "patient_data"  // BigQuery table
}
```
9. Create a staging table file `definitions/stg_patient_data.sqlx` with the following content:

```sql
config {type: "table"}
SELECT
  * EXCEPT(Birthdate,
    Age,
    Notes,
    Dose_mg),
  COALESCE(SAFE.PARSE_DATE('%m/%d/%Y', Birthdate),SAFE.PARSE_DATE('%Y-%m-%d', Birthdate)) Birthdate,
  JSON_VALUE(Notes, '$.Notes') Notes,
  CASE
    WHEN REGEXP_CONTAINS(Dose_mg, 'oz') THEN SAFE_CAST(SPLIT(Dose_mg, ' ')[OFFSET (0)] AS FLOAT64) * 28349.5231
    WHEN REGEXP_CONTAINS(Dose_mg, 'mg') THEN SAFE_CAST(SPLIT(Dose_mg, ' ')[OFFSET (0)] AS FLOAT64)
    ELSE NULL
  END Dose_mg
FROM
  ${ref("patient_data")}
```

10. Create a javascript file `includes/functions.js`

```javascript
function convert_to_mg(dose) {
    return `
        CASE
            WHEN REGEXP_CONTAINS(${dose}, 'oz') THEN SAFE_CAST(SPLIT(${dose}, ' ')[OFFSET(0)] AS FLOAT64) * 28349.5231
            WHEN REGEXP_CONTAINS(${dose}, 'mg') THEN SAFE_CAST(SPLIT(${dose}, ' ')[OFFSET(0)] AS FLOAT64)
            ELSE
                NULL
            END`;
}

module.exports = {
    convert_to_mg
};
```

11. Change the table file `definitions/patient_data.sqlx` to:

```sql
config {type: "table"}
SELECT
  * EXCEPT(Birthdate,
    Age,
    Notes,
    Dose_mg),
  COALESCE(SAFE.PARSE_DATE('%m/%d/%Y', Birthdate),SAFE.PARSE_DATE('%Y-%m-%d', Birthdate)) Birthdate,
  JSON_VALUE(Notes, '$.Notes') Notes,
  ${functions.convert_to_mg("Dose_mg")} Dose_mg
FROM
  ${ref("patient_data")}
```

12. Create javasctipt file `includes/constants.js` with the content:

```javascript
const infant = 1;
const toddler = 4;
const preschool = 6;
const gradeschhool = 12;
const teenager = 20;

module.exports = {
    infant,
    toddler,
    preschool,
    gradeschhool,
    teenager
};
```

13. Add the following code to `includes/functions.js`:

```javascript
function age_segment(age) {
    return `
        CASE
            WHEN ${age} < ${constants.infant} THEN "infant"
            WHEN ${age} < ${constants.toddler} THEN "toddler"
            WHEN ${age} < ${constants.preschool} THEN "preschool"
            WHEN ${age} < ${constants.gradeschhool} THEN "gradeschhool"
            WHEN ${age} < ${constants.teenager} THEN "teenager"
        ELSE
            "adult"
        END`;
}
```
and change `module.exports` to
```javascript
module.exports = {
    convert_to_mg,
    age_segment
};
```

11. Create a view file  `definitions/patient_info.sqlx` with the content:

```sql
config { type: "view" }
WITH
  t AS (
  SELECT
    DATE_DIFF(CURRENT_DATE('UTC'), Birthdate, YEAR) AS Age
  FROM
    ${ref("stg_patient_data")} )
SELECT
  *,
  ${functions.age_segment("Age")} AS age_segment
FROM
  t
```

15. Create an assertion file `definitions/patient_assertion.sqlx` with the content:

```sql
config { type: "assertion" }
SELECT
  *
FROM
  ${ref("stg_patient_data")}
WHERE
  NOT REGEXP_CONTAINS(Height, '^\\d+(\\.\\d*)? cm$')
```

16. Change the config section in `definitions/patient_info.sqlx` to:

```json
config {
    type: "view",
    dependencies: [ "patient_assertion"]
}
```

17. Execute the workflow, it should fail due to an assertion exception.

18. Add the follwoing function to `includes/functions.js`:

```javascript
function convert_to_cm(height) {
    return `
        CASE
            WHEN REGEXP_CONTAINS(${height}, 'ft') THEN CONCAT(ROUND((SAFE_CAST(SPLIT(${height}, ' ')[OFFSET(0)] AS FLOAT64) * 30.48),2)," cm")
            WHEN REGEXP_CONTAINS(${height}, 'cm') THEN ${height}
            ELSE
                NULL
            END`;
}
```
and change `module.exports` to:
```javascript
module.exports = {
    convert_to_cm,
    convert_to_mg,
    age_segment
};
```

19. Add the new function to `definitions/stg_patient_data.sqlx`:

```sql
config {type: "table"}
SELECT
  * EXCEPT(Birthdate,
    Age,
    Notes,
    Dose_mg,
    Height),
  COALESCE(SAFE.PARSE_DATE('%m/%d/%Y', Birthdate),SAFE.PARSE_DATE('%Y-%m-%d', Birthdate)) Birthdate,
  JSON_VALUE(Notes, '$.Notes') Notes,
  ${functions.convert_to_mg("Dose_mg")} Dose_mg,
  ${functions.convert_to_cm("Height")} Height
FROM
  ${ref("patient_data")}
```

20. Execute the workflow again, now it should run successfully

```json
config {
  type:   "declaration",
  schema: "dataform_ds",  // BigQuery dataset
  name:   "product_data"  // BigQuery table
}
```

2. Create a new incremental table in `definitions/product_inc.sqlx`:

```sql
config {
    type: "incremental",
}

SELECT
  *
FROM
  ${ref("product_data")}
  ${
      when(
          incremental(),
          `WHERE ingestion_time > (SELECT MAX(ingestion_time)
           FROM ${self()})`
      )
  }
```
